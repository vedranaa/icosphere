# Icosphere

Showing how module `icosphere` is to be used to generate geodesic icosahedron for a given subdivision frequency `nu`.

```python
import icosphere
import plotly.graph_objects as go
from plotly.subplots import make_subplots


def mesh_plot(vertices, faces):
    x, y, z = vertices.T
    i, j, k = faces.T
    return go.Mesh3d(x=x, y=y, z=z, i=i, j=j, k=k)


def wireframe_plot(vertices, faces):
    # Collect unique edges from all triangular faces.
    edges = set()
    for a, b, c in faces:
        edges.add(tuple(sorted((a, b))))
        edges.add(tuple(sorted((b, c))))
        edges.add(tuple(sorted((c, a))))

    x, y, z = [], [], []
    for i, j in edges:
        x.extend([vertices[i, 0], vertices[j, 0], None])
        y.extend([vertices[i, 1], vertices[j, 1], None])
        z.extend([vertices[i, 2], vertices[j, 2], None])

    line_style = dict(color="rgb(40,40,40)", width=1)
    return go.Scatter3d(x=x, y=y, z=z, mode="lines", line=line_style, showlegend=False)


nu = 15
vertices, faces = icosphere.icosphere(nu)

fig = go.Figure()
fig.add_trace(mesh_plot(vertices, faces))
fig.add_trace(wireframe_plot(vertices, faces))

fig.update_layout(title_text="Icosphere", height=600, width=600)
fig.show()
```

```python
fig = make_subplots(
    rows=2,
    cols=3,
    specs=[[{"type": "surface"}, {"type": "surface"}, {"type": "surface"}],
           [{"type": "surface"}, {"type": "surface"}, {"type": "surface"}]],
)

for idx, nu in enumerate(range(1, 7)):
    row = idx // 3 + 1
    col = idx % 3 + 1
    vertices, faces = icosphere.icosphere(nu)
    fig.add_trace(mesh_plot(vertices, faces), row=row, col=col)
    fig.add_trace(wireframe_plot(vertices, faces), row=row, col=col)

fig.update_layout(title_text="Different values of nu", height=600, width=800, showlegend=False)
fig.show()
```
